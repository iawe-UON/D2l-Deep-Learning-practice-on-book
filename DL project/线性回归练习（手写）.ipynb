#### 练习：线性回归模型

```python
%matplotlib inline
import torch 
import random
import pandas as pd
import numpy as np
from d2l import torch as d2l
from torch.utils import data
```

```python
true_w=torch.tensor([2,3.4])
true_b=4.2
num_samples=10000
```

```python
def synthetic_data(num_sample,w,b):#生成数据集
    x=torch.normal(0,1,(num_sample,len(w)))
    y=torch.matmul(x,w)+b
    y+=torch.normal(0,0.1,y.shape)
    return x,y.reshape(-1,1)
```

```python
features,labels=synthetic_data(num_samples,true_w,true_b)
features,labels
```

```python
def data_loader(batch_size,num_sample,feature,label):#生成小批量数据
    indice=list(range(num_sample))
    random.shuffle(indice)
    for i in range(0,num_sample,batch_size):
        dataset=torch.tensor(indice[i:min(i+batch_size,num_sample)])
        yield feature[dataset],label[dataset]
```

```python
batch_size=100
for x,y in data_loader(batch_size,num_samples,features,labels):
    print(x,'\n',y)
    break
```

```python
def linear(x,w,b):
    return torch.matmul(x,w)+b
def loss(y,y_hat):
    return (y_hat-y.reshape(y_hat.shape))**2/2
def sgd(batch_size,lr,params):
    with torch.no_grad():
        for param in params:
            param-=lr*param.grad/batch_size
            param.grad.zero_()
```

```python
"""初始化"""
w=torch.normal(0,0.5,size=(2,1),requires_grad=True)
b=torch.zeros(1,requires_grad=True)
```

```python
"""训练"""
num_epoch=
lr=0.01
for epoch in range(num_epoch):
    for x,y in data_loader(batch_size,num_samples,features,labels):
        y_hat=linear(x,w,b)
        l=loss(y,y_hat)
        l.sum().backward()
        sgd(batch_size,lr,[w,b])
    with torch.no_grad():
        train_loss=loss(labels,linear(features,w,b))
        print(f'epoch={epoch+1},loss={float(train_loss.mean()):.5f}')
```
