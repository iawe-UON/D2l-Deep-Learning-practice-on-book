```python
%matplotlib inline
import torch
from torch import nn
from d2l import torch as d2l
import numpy as np
```

```python
#生成一串有正态噪音的sin序列
t=1000
x=torch.arange(t,dtype=torch.float32)
y=torch.normal(0,0.1,(t,))+torch.sin(0.01*x)
d2l.plot(x,[y],'time','y',xlim=[0,t],figsize=(6,3))
```

```python
tau=4
feature=torch.zeros((t-tau,tau))
for i in range(tau):
    feature[:,i]=y[i:t-tau+i]
labels=y[tau:].reshape((-1,1))
batch_size=16
n_num=600
train_iter=d2l.load_array((feature[:n_num],labels[:n_num]),batch_size,is_train=True)
```

```python
def init_weight(m):
    if type(m)==nn.Linear:
        nn.init.xavier_uniform_(m.weight)
net=nn.Sequential(
    nn.Linear(4,10),
    nn.ReLU(),
    nn.Linear(10,1)
)
net=net.apply(init_weight)
loss=nn.MSELoss()
```

```python
num_epoch,lr=5,0.01
def train_ch3(net,train_iter,lr,num_epoch,batch_size,loss):
    for epoch in range(num_epoch):
        optimizer=torch.optim.Adam(net.parameters(),lr)
        net.train()
        for X,y in train_iter:
            optimizer.zero_grad()
            y_hat=net(X)
            l=loss(y_hat,y)
            l.sum().backward()
            optimizer.step()
        print(f'epoch{epoch+1},loss{d2l.evaluate_loss(net,train_iter,loss):.3f}')
train_ch3(net,train_iter,lr,num_epoch,batch_size,loss)
```

## 自回归模型：单步预测

```python
out=net(feature)
d2l.plot([x,x[tau:]],[y.detach().numpy(),out.detach().numpy()],'x','y',legend=['data','1-step-pre'],xlim=[1,1000],figsize=(6,3))
```
